Numbers (int & float)

```python
import math

print(type (1.0))

print("When comparing floats directly, we may run into precision issues")
print("0.1 * 3 == 0.3: ", 0.1 * 3 == 0.3 )
print("To tackle this we can use the math.isclose() function:")
print("math.isclose(0.1 * 3, 0.3): ", math.isclose(0.1 * 3, 0.3))
math.isclose(0.1 * 3, 0.3) # checks if the product of 0.1 * 3 is equal to 0.3
```

Arithmetic Operations

```python
print(type(4/2))
print(5/3)
print(type(5//3))
print(5//3)
print(5//3.0)
print(type(5//3.0))
print(5%3) # 1 as the result, 2 as remainder
```
