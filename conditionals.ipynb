## If / Elif / Else logic

Controls the flow of scripts based on conditions using `if`, `elif` and `else`.

## The `if` statement

An `if` statement executes a block of code only if a condition is `True`.

* Syntax: `if<condition>:` followed by an indented block
* Comparison operators: `==`, `!=`, `<`, `>`, `<=`, '>=`, `in`
* Combine conditions with `and`, `or`, `not`

```python
server_status = "running"

if server_status == "running":
    print("service is active")

if "r" in server_status:
    print("service is active")

if server_status:
    print("service is active")
```

## Truthiness

Python treates many values as truthy or falsy in conditionals.

* **Falsy:** `False`, `None`, `0`, `0.0`, `''`, `[]`, `{}`
* **Truthy:** non-zero numbers, non-empty sequences/collections

```python
servers = ["web01", "web02"]
error_message = "" # an empty string will be taken as null/false
default_config = {} # an empty dictionary is taken as null/false

if servers:
    print(f"Processing {len(servers)} servers.")

if error_message:
    print("Something went wrong:", error_message)

if not default_config:
    print("Default config not available, please provide the configuration values.")
```

## The else statement

Use `else` to execute code when the `if` condition is false.

```python
cpu_usage = 85.0

if cpu_usage > 90.0:
    print("ALERT: High CPU Usage")
else:
    print("CPU Usage is normal")
```

## The `elif` statement

Chain multiple checks, the first true block runs.

```python
https_status = 403

if https_status == 200:
    print("Status OK")
elif https_status == 404:
    print("Resource not found")
elif https_status >= 500:
    print("Server error (5xx)")
else:
    print("Another status:", https_status)
```

## Guard Clauses

Handle edge cases at the top of functions to avoid deep nesting of `if` conditions.

```python
def process_data_gaurded(data):
    if not data:
        print("No data provided")
    elif not isinstance(data, list):
        print(f"Invalid value type for 'data'. Provded {type(data)}; Required: list")
    else:
        print(f"Processing {len(data)} items...")
        print("Processed")

process_data_gaurded(None)
process_data_gaurded([])
process_data_gaurded("abc")
process_data_gaurded(10)
process_data_gaurded([1, 2, 3])
```
