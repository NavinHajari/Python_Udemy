Common Operations & Essential String Methods

1. Concatenation (+) : Join strings.
2. Length ( len() ) : Gets the number of characters.
3. Indexing ( [] ) : Access a character by position (0-based).
4. Slicing ( [:] ) : Extract substrings.
5. .lower() / .upper()
6. .strip() / .lstring() / ,rstrip()
7. .startswith() / .endswith()
8. .replace()

```python
course_title = "     Python for Devops     "
print(course_title)
print(f"Result of .strip(): {course_title.strip()}")
print(f"Result of .strip(): {"     Python for Devops     ".strip()}")
print(f"Result of .lstrip(): {course_title.lstrip()}") # removes the leading whitespaces
print(f"Result of .rstrip(): {course_title.rstrip()}") # removes the trailing whitespaces
print(f"Result of .upper(): {course_title.upper()}")
print(f"Result of .lower(): {course_title.lower()}")

filename = "file.yaml"
print(filename.startswith("file"))
print(filename.endswith("yaml"))

path = "/usr/local/bin"
path_parts = path.split("/")
print(f"path_parts: {path_parts}")
print(f"joined path paths: {"\\".join(path_parts)}")
print(path+"/python") #concatenation
print(len(path)) # prints the length of the string

# Slicing a string
print(path[1])
print(path[3:10]) # accessing specific characters from a string, prints from 3rd character to 9th (10th char is not printed)
print(path[:10]) # prints 0th to 9th chars
```

Exercise: Calculate Disk Usage Percentage

```python
server_name = "webserver-03"
cpu_cores = 4
memory_gb = 8.0
disk_total_gb = 500
disk_used_gb = 350

disk_usage_percentage = disk_used_gb / disk_total_gb
print (disk_usage_percentage)

summary = f"Server '{server_name.upper()}' ({cpu_cores} cores, {memory_gb}GB RAM) Disk Usage: {disk_usage_percentage:.3f})"
print(summary)

summary_formatted = f"Server '{server_name.upper()}' ({cpu_cores} cores, {memory_gb}GB RAM) Disk Usage: {disk_usage_percentage:.2%})"
print(summary_formatted)

```
