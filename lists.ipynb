Characteristics and Use Cases
1. Ordered items maintain position
2. mutable .append(), insert(), pop(), .remove()
3. Ideal for storing sequences where oder matters & contents change (e.g., list of servers, deployment steps)

Accessing items & Slicing


```python
servers = ["web01", "web02", "web03"]
mixed_list = ["config.yaml", 8080, True]

for item in mixed_list:
    print(type(item))

print(servers[0]) # returns 1st element
print(servers[2]) # returns 3rd element
print(servers[-1]) # returns last element
print(servers[-2]) # returns last but one 
# print(servers.[3]) # returns IndexError as it exceeds the limit
print(len(servers)) # returns the length of the list

# Slicing
print(servers[:2]) # Returns elements at indices 0 & 1
print(servers[:2]) # Returns elements at indices 0 & 1
print(servers[1:]) # Returns elements at indices 1 & 2
print(servers[-2:]) # Returns only second to last & last elements 

#Slicing doesn't alter the original list

print(servers)
```

```python
# Mutating lists
# servers.remove? # returns the function of .remove & its use
ports1 = [80, 443, 8080]
ports1.append(500)
print(ports1)
ports1.insert(1,600)
print(ports1)
ports1.remove(80)
print(ports1)
removed_value = ports1.pop(1)
print(ports1)
print(removed_value)
print(ports1)

def mutate_list(l):
    l.pop()

new_list = ["a", "b", "c"]
mutate_list(new_list)
print(new_list)
```
