## Efficient Looing: `range`

- Creating large lists for loops is memory-intensive (e.g., `list(range(1_000_000))`).
- `range()` stores only start, stop and step values and not all numbers.
- Numbers are generated one at a time during the iteration, reducing memory usage.
- Ideal for loops needing a fixed number of literations without large allocations.

```python
import sys

number_count = 10_000_00

numbers_list = list(range(number_count))
numbers_range = range(number_count)

print(numbers_list[:20])        # list readily stores the numbers in given range taking up more memory
print(numbers_range[:20])       # Whereas range only stores the start, stop and steps values taking up less memory

#for number in numbers_range[:20]:
#    print(number)

list_mb = sys.getsizeof(numbers_list) / (1024**2)
range_mb = sys.getsizeof(numbers_range) / (1024**2)

print(f"List size: {list_mb:.2f}")
print(f"Range size: {range_mb:.6f}")
print(f"List uses {(list_mb / range_mb):.3f} more memory.")
```

## Using `range()`
- range(stop): iterates from 0 up to (but not including) stop.
- range(start, stop): iterate from start up to stop
- range(start, stop, step): iterate with a custom stop increment.

```python
for i in range(5):                       # start defaults to '0' and range extends upto but not including '5'
    print(f"Retry: #{i}")

for year in range(2020, 2024):           # start sets to year '2020' and range extends upto but not including '2024'
    print(f"Processing logs for {year}")

for server_id in range(10, 30, 5):       # start sets to '10' and extends upto to but not including 30, in steps of '5'
    print(f"Checking server {server_id}")
```

## Getting Index + Value: `enumerate()`

- Use `enumerate(iterable, start=0)` to get `(index, item)` tuples.
- The start parameter sets the initial index value.

```python
servers = ["web01", "web02", "web03"]

for idx, server in enumerate(servers):
    print(f"#{idx}: Processing server {server}")
```

## Parallel Iteration: `zip()`

- Use `zip(*iterables)` to pair items from multiple iterables.
- Iteration stops when the shortest iterable is exhausted.

```python
hosts = ["hostA", "hostB", "hostC"]
ips = ["10.0.0.1", "10.0.0.2"]
azs = ["us-east-1a", "us-east-1b"]

for host, ip, az in zip(hosts, ips, azs):
    print(f"Host: {host}, IP: {ip}, AZ: {az}")
```
